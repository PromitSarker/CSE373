```python
import sys
import time
import matplotlib.pyplot as plt
import random
# For the brute force solution of the problem,  used top-down recursive procedure.

def rod_cutting(price, length):
    if length <= 0:
        return 0
    max_val = -sys.maxsize -1
    for i in range(length):
        max_val = max(max_val, price[i] + rod_cutting(price, length - (i + 1)))
        # continuous loop till length is equal or less than 0. Will fetch the max price and store it into max_val. 
    return max_val 

# length = int( input("Enter rod's length : "))
# price = []
# print("Input the value of cuttings : ")
# for j in range(0, length):
#     val = int( input())
#     price.append(val)
# price = [3, 5, 8, 9, 10, 17, 17, 20]
# length = 8

times_brute_force = []

sizes = list(range(1, 20, 2))

for length in sizes:
        price = [random.randint(1, 100) for _ in range(length)]  # Generate random price list
        start_time = time.time()
        rod_cutting(price, length-1)  # Compute the optimal price for a length
        end_time = time.time()
        execution = end_time - start_time
        times_brute_force.append(execution)
        
plt.plot(sizes, times_brute_force, label="Brute Force",marker='x')
plt.xlabel("Rod Length")
plt.ylabel("Time (seconds)")
plt.title("Time Complexity of Rod Cutting Algorithms")
plt.grid(True)
plt.legend()
plt.show()


# print("Maximum obtainable value : " , rod_cutting(price, length))


```
