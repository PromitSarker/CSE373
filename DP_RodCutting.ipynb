```python
import random
import time
import matplotlib.pyplot as plt

INFINITY_VAL = 100000

def find_max(x, y):
    if x > y:
        return x
    return y

def dynamic_rod_cutting(prices, rod_length):
    optimal_prices = [0] * (rod_length + 1)
    optimal_prices[0] = 0

    for j in range(1, rod_length + 1):
        max_revenue = -1 * INFINITY_VAL
        for i in range(1, j + 1):
            if max_revenue < prices[i] + optimal_prices[j - i]:
                max_revenue = prices[i] + optimal_prices[j - i]
        optimal_prices[j] = max_revenue

input_sizes = list(range(100, 10001, 1000))
execution_times = []

for rod_length in input_sizes:
    price_list = [random.randint(1, rod_length) for _ in range(rod_length)]
    start_time = time.time()
    dynamic_rod_cutting(price_list, rod_length - 1)
    end_time = time.time()
    execution_time = end_time - start_time
    execution_times.append(execution_time)

plt.plot(input_sizes, execution_times, marker='o')
plt.xlabel('Input Size (Amount)')
plt.ylabel('Execution Time (Seconds)')
plt.title('Dynamic Programming Approach to Rod Cutting: Average Case Time Complexity')
plt.show()
```
